```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
import pprint as pp
from key import api_key 

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Set base url
url = "http://api.openweathermap.org/data/2.5/weather?"

# Create counter to track record processing
counter = 0

# Create arrays to store city weather data
city_data = []
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lon = []
max_temp = []
wind_speed = []

print("Beginning Data Retrieval")     
print("-----------------------------")

# Loop through all cities
for city in cities:
        
        try:
            # Build query url for current city
            query_url = url + "appid=" + api_key + "&units=imperial" + "&q=" + city
            
            # Request data from openweathermap api
            weather_response = requests.get(query_url).json()
            
            # Append output data to arrays
            city_name.append(weather_response['name'])
            cloudiness.append(weather_response['clouds']['all'])
            country.append(weather_response['sys']['country'])
            date.append(weather_response['dt'])
            humidity.append(weather_response['main']['humidity'])
            lat.append(weather_response['coord']['lat'])
            lon.append(weather_response['coord']['lon'])
            max_temp.append(weather_response['main']['temp_max'])
            wind_speed.append(weather_response['wind']['speed'])
            
            #Add to counter
            counter = counter + 1
            
            #Print record number as processing
            print(f"Processing Record {counter} of Set 1 | {city}")
        
        # If there is no record for that city, display error and continue running code
        except KeyError:
            print('City not found. Skipping...')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Put all data into dataframe
cities_df = pd.DataFrame({"City": city_name,
                          "Cloudiness": cloudiness,
                          "Country": country,
                          "Date": date,
                          "Humidity": humidity,
                          "Lat": lat,
                          "Lon": lon,
                          "Max Temp": max_temp,
                          "Wind Speed": wind_speed})
cities_df.count()
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames
url = "http://api.openweathermap.org/data/2.5/weather?"

#Northern hemisphere arrays
city_data_north = []
city_name_north = []
cloudiness_north = []
country_north = []
date_north = []
humidity_north = []
lat_north = []
lon_north = []
max_temp_north = []
wind_speed_north = []

#Southern hemisphere arrays
city_data_south = []
city_name_south = []
cloudiness_south = []
country_south = []
date_south = []
humidity_south = []
lat_south = []
lon_south = []
max_temp_south = []
wind_speed_south = []

print("Beginning Data Retrieval")     
print("-----------------------------")

for city in cities:
        
        try:
            
            # Pull weather data from API
            query_url = url + "appid=" + api_key + "&units=imperial" + "&q=" + city
            weather_response = requests.get(query_url).json()
            
            #Assign variable for lat_coordinate
            lat_coordinate = weather_response['coord']['lat']
            
            #Create if function, pull only northern hemisphere data if lat_coordinate is greater than 0
            if lat_coordinate >= 0:
                
                city_name_north.append(weather_response['name'])
                cloudiness_north.append(weather_response['clouds']['all'])
                country_north.append(weather_response['sys']['country'])
                date_north.append(weather_response['dt'])
                humidity_north.append(weather_response['main']['humidity'])
                lat_north.append(weather_response['coord']['lat'])
                lon_north.append(weather_response['coord']['lon'])
                max_temp_north.append(weather_response['main']['temp_max'])
                wind_speed_north.append(weather_response['wind']['speed'])
            
            #All other coordinates are southern hemisphere
            else:
                
                city_name_south.append(weather_response['name'])
                cloudiness_south.append(weather_response['clouds']['all'])
                country_south.append(weather_response['sys']['country'])
                date_south.append(weather_response['dt'])
                humidity_south.append(weather_response['main']['humidity'])
                lat_south.append(weather_response['coord']['lat'])
                lon_south.append(weather_response['coord']['lon'])
                max_temp_south.append(weather_response['main']['temp_max'])
                wind_speed_south.append(weather_response['wind']['speed'])
        
        except KeyError:
            
            print('City not found. Skipping...')
            
```

```python
#Create dataframes for northern and southern hemisphere data
northern_df = pd.DataFrame({"City": city_name_north,
                          "Cloudiness": cloudiness_north,
                          "Country": country_north,
                          "Date": date_north,
                          "Humidity": humidity_north,
                          "Lat": lat_north,
                          "Lon": lon_north,
                          "Max Temp": max_temp_north,
                          "Wind Speed": wind_speed_north})

southern_df = pd.DataFrame({"City": city_name_south,
                          "Cloudiness": cloudiness_south,
                          "Country": country_south,
                          "Date": date_south,
                          "Humidity": humidity_south,
                          "Lat": lat_south,
                          "Lon": lon_south,
                          "Max Temp": max_temp_south,
                          "Wind Speed": wind_speed_south})

northern_df.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Plot scatter
plt.scatter(lat_north, max_temp_north)

#Assign northern latitude values
lat_north = northern_df.loc[:, "Lat"]

#Run linear regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(lat_north, max_temp_north)

# Multiply slope by northern latitude data and add intercept to get regression values
regress_values = lat_north * slope + intercept

# Print predictive line equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot line of northern latitudes and their respective regression values
plt.plot(lat_north,regress_values,"g-")

# Add line equation to graph
plt.annotate(line_eq,(0,-10),fontsize=15,color="red")

# Plot title, x and y labels and r value
plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.title("Northern Hemisphere: City Latitude vs. Max Temp (03/14/20)")
print(f"The r-squared is: {rvalue}")
plt.savefig('Northern Hemisphere Regression: City Latitude vs. Max Temp.png')
```

```python
# Linear regression is modeling the relationship between Latitude and Max Temperature in the Northern Hemisphere.
# There is a strong negative correlation between Latitude and Max Temperature.
```
